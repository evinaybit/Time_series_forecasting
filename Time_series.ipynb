```python
prev=stock.iloc[:180]
test=stock.iloc[180:]
```

```python
def plott(prediction,title):
    plt.figure(figsize=(17,5))
    plt.plot(prev,label='Train')
    plt.plot(test,label='Actual')
    plt.plot(prediction,label='Predicted')
    plt.ylabel('Stock prices')
    plt.title(title,fontsize=20)
    plt.legend()
    plt.show()
```

```python
from sklearn.metrics import mean_squared_error as mse
```

```python
av=pd.Series(np.repeat(prev.mean(),68),index=test.index)
mse(av,test)
```

```python
plott(av,'Average')
```

```python
#weighted mean 1-180  1st row would be given the weight 1/180, 180th row will
#be weighted as 1
```

```python
weight=np.array(range(0,180))/180
weighted_train=np.multiply(prev,weight)
```

```python
weighted_average=sum(weighted_train)/sum(weight)
wav=pd.Series(np.repeat(weighted_average,68),index=test.index)
mse(wav,test)
```

```python
plott(wav,'weighted')
```
